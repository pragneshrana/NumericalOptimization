# 1.  Prove that among all triangles with a fixed perimeter $P$, the equilateral triangle is the one that maximizes the area.
REF: https://math.stackexchange.com/questions/592917/simple-proof-that-equilateral-triangles-have-maximum-area
perimeter of triangle is fixed. $a+b+c= p = 2s = const$

By Heron's formula, area of triangle is obtained as, <br>

$A = \sqrt{s(s-a)(s-b)(s-c)} $ <br>

where, $s=\frac{a+b+c}{2} = \frac{p}{2}$ and a,b,c indicates the length of the triangle sides. <br>

It can be written as, <br>
$c = 2s-a-b$

**Objective** is to for $\max A$ = $\max \sqrt{s(s-a)(s-b)(s-c)}$

now, to find condition for maximum area, differentiate with respect to a keeping other terms constant and then with b. <br>

$ \dfrac{dA}{da} = 0 $

$0 =\dfrac{d}{dt}\sqrt{s(s-a)(s-b)(s-c)} $ <br>

$0 =\dfrac{s(s-b)(2s-2a-b)}{2A}$ <br>

$0 =s(s-b)(2s-2a-b)$ <br>

As $b\neq s$ so, taking another solution (2s-2a-b)=0 <br>

Now, taking derivative w.r.t b, <br>

$ \dfrac{dA}{db} = 0 $

$0 =\dfrac{d}{dt}\sqrt{s(s-a)(s-b)(s-c)} $ <br>

$0 =\dfrac{s(s-a)(2s-2b-a)}{2A}$ <br>

$0 =s(s-a)(2s-2b-a)$ <br>

As $b\neq s$ so, taking another solution (2s-2a-b)=0 <br>

solving this equation gives,

$a=b=\frac{2s}{3}$ <br>

$c= 2s-a-b =\frac{2s}{3} = b =a $  <br>

**Thus proved that maximum area is possible when all sides are equal.**

**A qudrilateral Q is said to be cyclic if its all vertices lie on a circle.*** 


Show that: The cyclic quadrilateral Q has the largest area of all quadrilaterals with sides of the same length as those of Q.

Proof:
Let $x=cos \theta$ and $y=cos\phi$, 
where, $\theta$ and $\phi$ are in $(0,\pi)$.

By applying the law of cosines to the triangle ABC and ACD, wwe have $a^2+b^2-2abcx = c^2 + d^2 -2cdy$

$\frac{dy}{dx} = \frac{ab}{cd}$ ......(1)

Adding the areas of trianles ABC andf ACD the area of the quadrilateral is,

$K = \frac{1}{2} ab \sin \theta +  \frac{1}{2} cd \sin \phi = \frac{1}{2}(ab\sqrt{1-x^2}+cd\sqrt{1-y^2})$

using (1),

it follows that

$\dfrac{dK}{dx} = \frac{-ab}{2}(\frac{x}{\sqrt{1-x^2}}+\frac{y}{\sqrt{1-y^2}})$

so, $\frac{dK}{dx}=0$ if and only if $x=\pm y$

if $x=y$ then $\frac{dK}{dx}=0$ implies that x=0=y and $\theta = \pi/2 = \phi$ 

if $x=-y$ then $\cos \theta = -\cos \phi = \cos(\pi-\phi)$ and as $\theta$ and $\phi$ are in $[0,\pi]$ it can be written as, $\phi+\theta = \pi$ 

as,

$\frac{d^2}{dx^2} = \frac{-ab}{2} (\frac{1}{\sqrt{1-x^2}^3}+\frac{1}{\sqrt{1-y^2}^3}\frac{dy}{dx})$

and $\frac{dy}{dx} = \frac{ab}{cd} >0$ we have $\frac{d^2K}{dx^2}<0 \space \forall x \in [0,1]$ 

thus, area is maximized for quadrilateral when it is cyclic.

# 8. Prove the arithmetic mean geometric mean inequality, i.e., $$\dfrac{x_1+x_2+\cdots+x_n}n \geq \sqrt[n]{x_1x_2\cdots x_n}$$ where $x_i \geq 0$ and $n \in \mathbb{Z}^+$.


REF : https://math.stackexchange.com/questions/691807/proofs-of-am-gm-inequality

As, $(\sqrt{x_1}-\sqrt{x_2})^2 \geq 0$

it can be written that $\sqrt{x_1 x_2} \leq \frac{x_1+x_2}{2}$

Applying this inequality twice we can get,

$(x_1 x_2 x_3 x_4)^{\frac{1}{4}} \leq \frac{\sqrt{x_1 x_2}+\sqrt{x_3 x_4}}{2} \leq \frac{x_1+x_2+x_3+x_4}{4}$

By induction it can be shown that,

$(x_1 \cdots x_{2^k})^{\frac{1}{2^k}} \leq \frac{x_1+\ldots+x_{2^k}}{2^k} \space \forall k \geq 1$

To fill the gap between powers of two. 
so let $x_1,...,x_n$ be arbitary positive numbers and choose k such that $n \leq 2^k$

$\alpha_i := \begin{cases} x_i & i \leq n \\ A & n< i \leq 2^k \end{cases}$

where, $A:= \frac{x_1+\ldots+x_n}{n}$ using this we can get,

$\bigg( x_1 \ldots x_n A^{2^k-n} \bigg)^{\frac{1}{2^k}} \leq \frac{x_1+\ldots+x_n+(2^k-n) A}{2^k} = A.$

so,
$(x_1 \ldots x_n)^{1/n} \leq A = \frac{x_1+\ldots+x_n}{n}$
